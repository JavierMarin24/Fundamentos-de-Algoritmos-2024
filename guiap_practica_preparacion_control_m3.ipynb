Pregunta 1

```python
def leerDatos() :
    lista = []
    while True :
        titulo = input()
        if titulo == "FINAL" :
            break
        horas = int(input())
        minutos = int(input())
        segundos = int(input())
        vistas = int(input())
        tupla = (titulo.upper(), horas, minutos, segundos, vistas)
        lista.append(tupla)
    return lista

def mostrarUno(lista) :
    print(f"TOTAL VIDEOS LEÍDOS = {len(lista)}\n")
    print("DATOS VIDEOS")
    print("------------")
    for tupla in lista :
        print(f"TITULO : {tupla[0]} - DURACION {tupla[1]}:{tupla[2]}:{tupla[3]} - VISTAS : {tupla[4]}")

def calcularTiempoTotal(lista) :
    suma = 0
    for tupla in lista :
        segundosTotal = tupla[1] * 3600 + tupla[2] * 60 + tupla[3]
        suma += segundosTotal
    return suma

def sumarVistas(lista) :
    suma = 0
    for tupla in lista :
        suma += tupla[4]
    return suma

def menorTiempo(lista) :
    menorSegundos = lista[0][1] * 3600 + lista[0][2] * 60 + lista[0][3]
    for tupla in lista :
        sumaSegundos = tupla[1] * 3600 + tupla[2] * 60 + tupla[3]
        if sumaSegundos < menorSegundos :
            menorSegundos = sumaSegundos

    print(f"\nMENOR DURACIÓN EN SEGUNDOS = {menorSegundos}")
    print("\nTÍTULO(S) VIDEO(S) + VISTAS CON MENOR DURACIÓN EN SEGUNDOS")
    print("----------------------------------------------------------")
    for tupla in lista :
        sumaSegundos = tupla[1] * 3600 + tupla[2] * 60 + tupla[3]
        if sumaSegundos == menorSegundos :
            print(f"* {tupla[0]} - {tupla[4]} VISTAS")

def obtenerMas1000(lista) :
    nueva = []
    for tupla in lista :
        if tupla[4] > 1000 :
            nueva.append(tupla[0])
    return nueva

def mostrarLista(listaMas1000) :
    print(f"\nLISTA DE VIDEOS CON MÁS DE 1000 VISTAS ORDENADOS ALFABÉTICAMENTE")
    print(f"----------------------------------------------------------------")
    for titulo in listaMas1000 :
        print(titulo)

# --- MAIN ---
lista = leerDatos()
mostrarUno(lista)
tiempoTotalSeg = calcularTiempoTotal(lista)
horas = tiempoTotalSeg // 3600
tiempoTotalSeg = tiempoTotalSeg % 3600
minutos = tiempoTotalSeg // 60
segundos = tiempoTotalSeg % 60
print(f"\nTIEMPO TOTAL DE TODOS LOS VIDEOS {horas}:{minutos}:{segundos}")
sumaVisualizaciones = sumarVistas(lista)
promedio = round(sumaVisualizaciones / len(lista),1)
print(f"\nPROMEDIO DE VISTAS DE TODOS LOS VIDEOS = {promedio}")
menorTiempo(lista)
listaMas1000 = obtenerMas1000(lista)
largo = len(listaMas1000)
if largo == 0 :
    print("\nNO HAY VIDEOS CON MÁS DE 1000 VISUALIZACIONES")
else :
    listaMas1000.sort()
    mostrarLista(listaMas1000)
```

Pregunta 2

```python
def leerValidar() :
    while True :
        dato = int(input())
        if dato > 0 :
            return dato

def leerPartidas(N) :
    lista = []
    for k in range(N) :
        nombreJugador = input()
        personaje = input()
        mundo = int(input())
        nivel = int(input())
        monedas = int(input())
        tupla = (nombreJugador, personaje, mundo, nivel, monedas)
        lista.append(tupla)
    return lista

def mostrar(lista) :
    print("LISTA DE PARTIDAS")
    print("-----------------")
    for tupla in lista :
        print(tupla)

def mayorMonedas(lista) :
    mayor = lista[0][4]
    for tupla in lista :
        if tupla[4] > mayor :
            mayor = tupla[4]
    return mayor

def mostrarMayor(lista, masMonedas) :
    print("JUGADOR(ES) QUE RECOLECTARON MAYOR CANTIDAD DE MONEDAS")
    print("------------------------------------------------------")
    for tupla in lista :
        if tupla[4] == masMonedas :
            print(tupla[0])

def obtenerPersonajes(lista) :
    listaDos = []
    for tupla in lista :
        if tupla[1] not in listaDos :
            listaDos.append(tupla[1])
    listaTres = []
    for elemento in listaDos :
        contador = 0
        for tupla1 in lista :
            if elemento == tupla1[1] :
                contador += 1
        tupla2 = (elemento, contador)
        listaTres.append(tupla2)
    mayor = listaTres[0][1]
    for tupla in listaTres :
        if tupla[1] > mayor :
            mayor = tupla[1]
    listaCuatro = []
    for tupla in listaTres :
        if tupla[1] == mayor :
           listaCuatro.append(tupla[0])
    return listaCuatro

def obtenerMundos(lista) :
    listaDos = []
    for tupla in lista :
        if tupla[2] not in listaDos :
            listaDos.append(tupla[2])
    listaTres = []
    for elemento in listaDos :
        contador = 0
        for tupla1 in lista :
            if elemento == tupla1[2] :
                contador += 1
        tupla2 = (elemento, contador)
        listaTres.append(tupla2)
    mayor = listaTres[0][1]
    for tupla in listaTres :
        if tupla[1] > mayor :
            mayor = tupla[1]
    listaCuatro = []
    for tupla in listaTres :
        if tupla[1] == mayor :
           listaCuatro.append(tupla[0])
    return sorted(listaCuatro)

# MAIN
N = leerValidar()
lista = leerPartidas(N)
mostrar(lista)
masMonedas = mayorMonedas(lista)
print(f"\nLA MAYOR CANTIDAD DE MONEDAS FUE {masMonedas}")
mostrarMayor(lista, masMonedas)
personajesMasUsados = obtenerPersonajes(lista)
print(f"\nPERSONAJE(S) MÁS USADO(S) = {personajesMasUsados}")
mundosMasJugados = obtenerMundos(lista)
print(f"\nMUNDO(S) MÁS JUGADO(S) = {mundosMasJugados}")
```

```python
def poblar (num) :
    lista = []
    for aux in range(num) :
        dato = input()
        lista.append(dato)
    return lista

def esParitoso(num) :
    digitos = list(num)
    for dig in digitos :
        if int(dig) % 2 != 0 :
            return False
    return True

def filtrarParitosos(lista) :
    nueva = []
    for dato in lista :
        if esParitoso(dato) :
            nueva.append(int(dato))
    return nueva

############################

# MAIN

n = int(input())
listaDatos = poblar(n)
print("Lista inicial es:")
print(listaDatos)
print()
listaParitosos = filtrarParitosos(listaDatos)
listaParitosos.sort()
print("Lista de Paritosos es:")
print(listaParitosos)
print()
porcentaje = len(listaParitosos) / len(listaDatos) * 100
porcentaje = round(porcentaje,1)
print("Porcentaje de Paritosos es:",porcentaje,"%")

# FIN
```

```python
def leerDatos() :
    lista = []
    while True :
        cadena = input()
        if cadena == "FIN" : return lista
        if cadena.isdigit() and int(cadena) > 0 and int(cadena) not in lista :
            lista.append(int(cadena))

def sumarDivisores(numero) :
    suma = 0
    mitad = numero // 2
    for divisor in range(1, mitad + 1) :
        if numero % divisor == 0 :
            suma += divisor
    return suma

def obtenerListaTuplas(listaNumeros) :
    listaTuplas = []
    for numero in listaNumeros :
        sumaDiv = sumarDivisores(numero)
        tupla = (numero,sumaDiv)
        listaTuplas.append(tupla)
    return listaTuplas

def obtenerlistaAbundantes(listaTuplas) :
    listaAbundantes = []
    for tupla in listaTuplas :
        if tupla[1] > tupla[0]:
            listaAbundantes.append(tupla[0])
    return listaAbundantes

# --- MAIN ---
listaNumeros = leerDatos()
if listaNumeros == [] :
    print("LISTA INICIAL VACIA")
else :
    print(f"LISTA INICIAL = {listaNumeros}")
    print(f"TOTAL ELEMENTOS EN LISTA INICIAL = {len(listaNumeros)}")
    listaNumeros.sort(reverse = True)
    print(f"LISTA INICIAL ORDENADA DECRECIENTE = {listaNumeros}")
    listaTuplas = obtenerListaTuplas(listaNumeros)
    print(f"LISTA TUPLAS = {listaTuplas}")
    listaAbundantes = obtenerlistaAbundantes(listaTuplas)
    if listaAbundantes == [] :
        print("NO SE ENCONTRARON NÚMEROS ABUNDANTES EN LA LISTA")
    else :
        print(f"LISTA DE NÚMEROS ABUNDANTES = {listaAbundantes}")
        print(f"TOTAL ELEMENTOS EN LISTA DE NÚMEROS ABUNDANTES = {len(listaAbundantes)}")
        print("--- GRAN ABUNDANCIA ---")
```

```python
def leerDatos() :
    lista = []
    while True :
        cadena = input()
        if cadena == "FIN" : return lista
        if cadena.isdigit() and int(cadena) > 0 and int(cadena) not in lista :
            lista.append(int(cadena))

def sumarDigitos(numero) :
    suma = 0
    while numero != 0 :
        digito = numero % 10
        suma += digito
        numero //= 10
    return suma

def obtenerListaTuplas(listaNumeros) :
    listaTuplas = []
    for numero in listaNumeros :
        sumaDig = sumarDigitos(numero)
        tupla = (numero,sumaDig)
        listaTuplas.append(tupla)
    return listaTuplas

def obtenerListaHarshad(listaTuplas) :
    listaHarshad = []
    for tupla in listaTuplas :
        if tupla[0] % tupla[1] == 0 :
            listaHarshad.append(tupla[0])
    return listaHarshad

# --- MAIN ---
listaNumeros = leerDatos()
if listaNumeros == [] :
    print("LISTA INICIAL VACIA")
else :
    print(f"LISTA INICIAL = {listaNumeros}")
    print(f"TOTAL ELEMENTOS EN LISTA INICIAL = {len(listaNumeros)}")
    listaNumeros.sort()
    print(f"LISTA INICIAL ORDENADA CRECIENTE = {listaNumeros}")
    listaTuplas = obtenerListaTuplas(listaNumeros)
    print(f"LISTA TUPLAS = {listaTuplas}")
    listaHarshad = obtenerListaHarshad(listaTuplas)
    if listaHarshad == [] :
        print("NO SE ENCONTRARON NÚMEROS DE HARSHAD EN LA LISTA")
    else :
        print(f"LISTA DE NÚMEROS HARSHAD = {listaHarshad}")
        print(f"TOTAL ELEMENTOS EN LISTA DE NÚMEROS HARSHAD = {len(listaHarshad)}")
        print("--- GRAN ALEGRÍA ---")
```

```python
def leerDatos() :
    lista = []
    while True :
        cadena = input()
        if cadena == "FIN" : return lista
        if cadena.isdigit() and int(cadena) > 0 and cadena not in lista :
            lista.append(cadena)

def obtenerListaNumeros(listaStr) :
    listaNum = []
    for cadena in listaStr :
        listaNum.append(int(cadena))
    return listaNum

def sumarDigitos(numero,largo) :
    suma = 0
    while numero != 0 :
        digito = (numero % 10) ** largo
        suma += digito
        numero //= 10
    return suma

def obtenerListaTuplas(listaNumeros) :
    listaTuplas = []
    for numero in listaNumeros :
        largo = len(str(numero))
        sumaDig = sumarDigitos(numero,largo)
        tupla = (numero,largo,sumaDig)
        listaTuplas.append(tupla)
    return listaTuplas

def obtenerlistaNNarcisistas(listaTuplas) :
    listaNNarcisistas = []
    for tupla in listaTuplas :
        if tupla[0] == tupla[2] :
            listaNNarcisistas.append(tupla[0])
    return listaNNarcisistas

# --- MAIN ---
listaStr = leerDatos()
if listaStr == [] :
    print("LISTA INICIAL DE CADENAS VACIA")
else :
    print(f"LISTA INICIAL DE CADENAS = {listaStr}")
    print(f"TOTAL ELEMENTOS EN LISTA INICIAL DE CADENAS = {len(listaStr)}")
    listaNumeros = obtenerListaNumeros(listaStr)
    listaNumeros.sort()
    print(f"LISTA NUMEROS EN ORDEN CRECIENTE = {listaNumeros}")
    listaTuplas = obtenerListaTuplas(listaNumeros)
    print(f"LISTA TUPLAS = {listaTuplas}")
    listaNNarcisistas = obtenerlistaNNarcisistas(listaTuplas)
    if listaNNarcisistas == [] :
        print("NO SE ENCONTRARON NÚMEROS N-NARCISISTAS EN LA LISTA")
    else :
        print(f"LISTA DE NÚMEROS N-NARCISISTAS = {listaNNarcisistas}")
        print(f"TOTAL ELEMENTOS EN LISTA DE NÚMEROS N-NARCISISTAS = {len(listaNNarcisistas)}")
        print("--- GRAN NARCISISMO ---")
```
